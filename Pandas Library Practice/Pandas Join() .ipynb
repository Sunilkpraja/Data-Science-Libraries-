# Pandas Join() 

```python
import pandas as pd
```

```python
df1 = pd.DataFrame ({'A':[1,2,3,4],'B':[10,20,30,40]})
```

```python
df2 = pd.DataFrame ({'C':[5,6,7,8],'D':[50,60,70,80]})
```

```python
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1 = pd.DataFrame ({'A':[1,2,3,4],'B':[10,20,30,40]},index = ['a','b','c','d'])
df2 = pd.DataFrame ({'C':[5,6,7,8],'D':[50,60,70,80]},index = ['p','q','r','s'])
```

```python
display(df1,df2)
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'inner')
```

```python
df1.join(df2, how = 'outer')
```
