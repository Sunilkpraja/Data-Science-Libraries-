# Pandas concate() Function in Python

```python
import pandas as pd
```

```python
series1 = pd.Series([0,1,2,2,3])
series1
```

```python
series2 = pd.Series([3,4,5])
series2
```

```python
pd.concat([series1,series2])
```

```python
df1 = pd.DataFrame({'ID' : [1,2,3,4],
                       'Name' : ['A','B','C','D'],
                       'Class' :[5,6,7,8]})
```

```python
df2 = pd.DataFrame({'ID' : [5,6,7,8],
                       'Name' : ['P','Q','R','S'],
                       'Class' :[4,3,2,1]})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2, df1], axis = 1)
```

```python
pd.concat([df1, df2], axis = 0, ignore_index = True)
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

```python
pd.concat([df1, df2], axis = 1, join_axes = [df1.index])
```

```python
pd.concat([df1, df2], axis = 1, join_axes = [df2.index])
```

```python
pd.concat([df1, df2], keys = ['df1','df2'])
```

```python
pd.concat([df1, df2], keys = ['First df','Second df'])
```

```python
pd.concat([df1, df2], axis = 1,  keys = ['First df','Second df'])
```

```python
pd.concat([df1, df2], sort = False)
```
